## **Handling Missing Data**

```python
# Check for missing values
missing_values = train_df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

# Display missing values
print("Missing Values Per Column:\n", missing_values)
```

### Approach for Handling Missing Values:
- Categorical Features:
  - Fill missing laundry_options and parking_options with "unknown" to preserve information without introducing bias.
- Numerical Features:
  - Fill missing lat and long with the median of their respective region to maintain geographical accuracy.
  - Compute price_per_sqft based on price and sqfeet where possible; otherwise, use median price per sqft from similar properties

```python
# Filling categorical missing values
train_df["laundry_options"].fillna("unknown", inplace=True)
train_df["parking_options"].fillna("unknown", inplace=True)

# Fill missing lat/long using regional median values
train_df["lat"] = train_df["lat"].fillna(train_df.groupby("region")["lat"].transform("median"))
train_df["long"] = train_df["long"].fillna(train_df.groupby("region")["long"].transform("median"))

# Compute price_per_sqft where missing
train_df["price_per_sqft"] = train_df["price"] / train_df["sqfeet"]
median_price_sqft = train_df.groupby("region")["price_per_sqft"].median()

# Apply median price per sqft for missing values
train_df.loc[train_df["price_per_sqft"].isnull(), "price_per_sqft"] = train_df["region"].map(median_price_sqft)

# Verifying if missing values are handled
missing_values_after = train_df.isnull().sum()
print("Missing Values After Cleaning:\n", missing_values_after[missing_values_after > 0])
```

Since the description column is unlikely to be used directly in our ML pipeline, we can safely replace missing values with "No description provided" to maintain dataset consistency.

```python
# Fill missing descriptions with a placeholder text
train_df["description"].fillna("No description provided", inplace=True)

# Verify that no missing values remain
missing_values_final = train_df.isnull().sum()
print("Final Missing Values:\n", missing_values_final[missing_values_final > 0])
```

## **Feature Engineering & Transformation**
To improve the model’s predictive power, I’ll create **new features** and transform existing ones to ensure they are **model-ready**. This includes:

-	Deriving meaningful **property-specific metrics** like `price_per_sqft`.
-	Encoding **categorical variables** to numerical format.
-	Scaling numerical features to **standardize values** across different scales.
-	Ensuring that feature selection aligns with **business needs**.


```python
# Feature: Price per Bedroom
train_df["price_per_bed"] = train_df["price"] / train_df["beds"]
train_df["price_per_bed"].replace([np.inf, -np.inf], np.nan, inplace=True)
train_df["price_per_bed"].fillna(train_df["price"].median(), inplace=True)

# Feature: Price per Bathroom
train_df["price_per_bath"] = train_df["price"] / train_df["baths"]
train_df["price_per_bath"].replace([np.inf, -np.inf], np.nan, inplace=True)
train_df["price_per_bath"].fillna(train_df["price"].median(), inplace=True)

# Feature: Pets Allowed Indicator
train_df["pets_allowed"] = (train_df["cats_allowed"] | train_df["dogs_allowed"]).astype(int)

# Feature: Total Amenities Count
train_df["total_amenities"] = (
    train_df["pets_allowed"] + train_df["smoking_allowed"] +
    train_df["wheelchair_access"] + train_df["electric_vehicle_charge"] +
    train_df["comes_furnished"]
)

# Feature: Number of Listings per Region
train_df["region_listings"] = train_df.groupby("region")["id"].transform("count")

# Display sample data
print("Feature Engineering Completed. Sample Data:")
display(train_df[["price_per_bed", "price_per_bath", "pets_allowed", "total_amenities", "region_listings"]].head())


```

```python
# Ensure no zero values before division
test_df["sqfeet"] = test_df["sqfeet"].replace(0, np.nan)
test_df["beds"] = test_df["beds"].replace(0, np.nan)
test_df["baths"] = test_df["baths"].replace(0, np.nan)

# Compute missing features
test_df["price_per_sqft"] = test_df["price"] / test_df["sqfeet"]
test_df["price_per_bed"] = test_df["price"] / test_df["beds"]
test_df["price_per_bath"] = test_df["price"] / test_df["baths"]
test_df["region_listings"] = test_df.groupby("region")["id"].transform("count")

# Handle NaN values by filling with median
test_df["price_per_sqft"].fillna(test_df["price_per_sqft"].median(), inplace=True)
test_df["price_per_bed"].fillna(test_df["price_per_bed"].median(), inplace=True)
test_df["price_per_bath"].fillna(test_df["price_per_bath"].median(), inplace=True)
test_df["region_listings"].fillna(test_df["region_listings"].median(), inplace=True)


# Confirm Fix
print("Fixed Missing Features in Test Set:")
display(test_df[["price_per_sqft", "price_per_bed", "price_per_bath", "region_listings"]].head())
```

```python
# Initialize XGBoost model
xgb_model = xgb.XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=6, random_state=42)

# Train the model
xgb_model.fit(X_train, y_train)

# Predict on test set
y_pred_xgb = xgb_model.predict(X_test)

# Evaluate performance
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2_xgb = r2_score(y_test, y_pred_xgb)

# Print performance metrics
print("XGBoost Model Performance:")
print(f"Mean Absolute Error (MAE): {mae_xgb:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_xgb:.2f}")
print(f"R-Squared (R²): {r2_xgb:.4f}")
```

```python
xgb_tuned = xgb.XGBRegressor(
    n_estimators=500,          # Increase number of trees
    learning_rate=0.01,        # Reduce learning rate
    max_depth=8,               # Increase tree depth
    subsample=0.8,             # Use only 80% of data per tree
    colsample_bytree=0.8,      # Use only 80% of features per tree
    reg_alpha=0.1,             # L1 Regularization (Feature Selection)
    reg_lambda=0.1,            # L2 Regularization (Prevent Overfitting)
    random_state=42
)

# Train XGBoost with tuned hyperparameters
xgb_tuned.fit(X_train, y_train)

# Predict on test set
y_pred_xgb_tuned = xgb_tuned.predict(X_test)

# Evaluate performance
mae_xgb_tuned = mean_absolute_error(y_test, y_pred_xgb_tuned)
rmse_xgb_tuned = np.sqrt(mean_squared_error(y_test, y_pred_xgb_tuned))
r2_xgb_tuned = r2_score(y_test, y_pred_xgb_tuned)

# Print performance metrics
print("Tuned XGBoost Model Performance:")
print(f"Mean Absolute Error (MAE): {mae_xgb_tuned:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_xgb_tuned:.2f}")
print(f"R-Squared (R²): {r2_xgb_tuned:.4f}")
```

### **Identifying the Issue**
Since XGBoost should **normally** outperform Linear Regression on complex datasets, we need to **diagnose why it's failing**:



#### **Re-check Feature Importance**
- **Are some features dominating too much?**
- **Are we using irrelevant or redundant features?**
- **Does feature importance suggest we should drop some features?**

```python


# Define numerical features (excluding one-hot encoded categorical features)
numerical_features = [
    "sqfeet", "beds", "baths", "price_per_sqft", "price_per_bed",
    "price_per_bath", "total_amenities", "region_listings"
]

# Compute mutual information scores
mi_scores = mutual_info_regression(train_df_encoded[numerical_features], train_df_encoded["price"])
mi_scores_series = pd.Series(mi_scores, index=numerical_features).sort_values(ascending=False)

# Compute correlation matrix
correlation_matrix = train_df_encoded[numerical_features].corr()

# Display results
print("Mutual Information Scores with Target (Price):")
print(mi_scores_series)

print("\nHighly Correlated Features (Threshold > 0.5):")
high_correlation_pairs = correlation_matrix.abs().stack().reset_index()
high_correlation_pairs = high_correlation_pairs[high_correlation_pairs[0] > 0.5]
high_correlation_pairs = high_correlation_pairs[high_correlation_pairs["level_0"] != high_correlation_pairs["level_1"]]
print(high_correlation_pairs)
```

### **Selecting Meaningful Feature Interactions**
Based on the mutual information (MI) scores and correlation analysis, we can make **data-driven** decisions about which interactions to create.

#### **Key Observations**
- **`price_per_bath` (MI: 5.99) and `price_per_bed` (MI: 5.93)** are the most important predictors of `price`.
- **`price_per_sqft` (MI: 4.01)** is also highly relevant.
- The three features (`price_per_sqft`, `price_per_bed`, and `price_per_bath`) are **highly correlated (>0.88)**, indicating they capture similar information.
- **`sqfeet`, `region_listings`, and `baths`** have moderate predictive power.
- **`beds` and `total_amenities`** have low MI scores, meaning they have a weaker relationship with `price`.

---

### **Decision: Which Interactions to Create?**
1. **Avoid Redundant Features**  
   - Since `price_per_sqft`, `price_per_bed`, and `price_per_bath` are strongly correlated, we **do not** need to create multiple interaction terms with them.  
   - Instead, we select **only one primary interaction** that represents size-based pricing.

2. **Create `price_sqft_region` (price_per_sqft × region_listings)**  
   - **Rationale:**  
     - `price_per_sqft` captures local rental pricing trends.  
     - `region_listings` represents supply density in the region.  
     - Multiplying them provides insight into whether **high-demand regions impact pricing per sqft**.

3. **Create `beds_baths` (beds × baths)**  
   - **Rationale:**  
     - Number of **bedrooms and bathrooms together** defines home capacity.  
     - A 1-bed/1-bath vs. 3-bed/2-bath home has different rental pricing power.
     - Multiplying them can enhance model differentiation.

```python
# Train a quick XGBoost model to check feature importance
xgb_model_temp = xgb.XGBRegressor(n_estimators=100, learning_rate=0.05, max_depth=6, random_state=42)
xgb_model_temp.fit(X_train, y_train)

# Get feature importance
plt.figure(figsize=(10, 5))
xgb.plot_importance(xgb_model_temp, importance_type="weight", max_num_features=20)
plt.title("Feature Importance (XGBoost)")
plt.show()
```

```python
# Define weak features to drop based on importance
drop_features = ["type_house", "state_5", "laundry_options_laundry on site", "lat", "long", "beds"]

# Drop from train & test sets
X_train = X_train.drop(columns=drop_features)
X_test = X_test.drop(columns=drop_features)

print("Updated Feature List:", X_train.columns.tolist())
```

```python
# Define the objective function
def objective(trial):
    params = {
        "n_estimators": trial.suggest_int("n_estimators", 100, 1000, step=100),
        "learning_rate": trial.suggest_loguniform("learning_rate", 0.005, 0.1),
        "max_depth": trial.suggest_int("max_depth", 3, 10),
        "subsample": trial.suggest_float("subsample", 0.6, 1.0),
        "colsample_bytree": trial.suggest_float("colsample_bytree", 0.6, 1.0),
        "reg_alpha": trial.suggest_loguniform("reg_alpha", 1e-4, 10.0),
        "reg_lambda": trial.suggest_loguniform("reg_lambda", 1e-4, 10.0),
    }

    model = xgb.XGBRegressor(**params, random_state=42, eval_metric="rmse")

    # Train the model and manually implement early stopping
    model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)

    # Make predictions
    y_pred = model.predict(X_test)

    # Return RMSE for evaluation
    return mean_squared_error(y_test, y_pred) ** 0.5 # RMSE

# Run optimization
study = optuna.create_study(direction="minimize")
study.optimize(objective, n_trials=20)

# Get best parameters
best_params = study.best_params
print("Best XGBoost Hyperparameters:", best_params)

# Train final model with best parameters
xgb_best = xgb.XGBRegressor(**best_params, eval_metric="rmse")
xgb_best.fit(X_train, y_train)
y_pred_xgb_best = xgb_best.predict(X_test)

# Evaluate performance
mae_xgb = mean_absolute_error(y_test, y_pred_xgb_best)
rmse_xgb = mean_squared_error(y_test, y_pred_xgb_best) ** 0.5
r2_xgb = r2_score(y_test, y_pred_xgb_best)

print(f"Fine-Tuned XGBoost Performance:\nMAE: {mae_xgb:.2f}, RMSE: {rmse_xgb:.2f}, R²: {r2_xgb:.4f}")
```
